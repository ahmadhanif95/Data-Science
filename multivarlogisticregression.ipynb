```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/content/heart.csv',encoding='latin1')
df.info()
df.isna().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.ChestPainType = le.fit_transform(dfle.ChestPainType)
dfle.RestingECG = le.fit_transform(dfle.RestingECG)
dfle.ExerciseAngina = le.fit_transform(dfle.ExerciseAngina)
dfle.Sex = le.fit_transform(dfle.Sex)
dfle.ST_Slope = le.fit_transform(dfle.ST_Slope)
dfle
X = dfle[['Age','ChestPainType','RestingBP','Cholesterol','RestingECG','MaxHR','ExerciseAngina',"Sex", "FastingBS","ST_Slope","Oldpeak"]]
y = dfle['HeartDisease']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
y_predicted = clf.predict(X_test)
```
