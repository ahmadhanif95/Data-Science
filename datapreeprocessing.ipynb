```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Import label encoder
from sklearn import preprocessing
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
```

```python
df = pd.read_csv('/content/Egypt_Houses_Price.csv')
```

```python
df1 = pd.read_csv('/content/insurance.csv')
```

```python
df = df[df['Bathrooms'] != 'Unknown']
df = df[df['Bedrooms'] != 'Unknown']
df = df[df['Area'] != 'Unknown']
df = df[df['Price'] != 'Unknown']
```

```python

df1 = df1[df1['bmi'] != 'Unknown']
df1 = df1[df1['children'] != 'Unknown']

df1 = df1[df1['charges'] != 'Unknown']
```

```python
df['Area'].replace('Unknown', np.nan)
df['Price'].replace('Unknown', np.nan)
df['Bedrooms'].replace('Unknown', np.nan)
df['Bathrooms'].replace('Unknown', np.nan)

```

```python
df1['bmi'].replace('Unknown', np.nan)
df1['children'].replace('Unknown', np.nan)
df1['charges'].replace('Unknown', np.nan)
```

```python
df['Bedrooms'] = df['Bedrooms'].replace('10+', '11')
df['Bathrooms'] = df['Bathrooms'].replace('10+', '11')
```

```python
df['Area'] = df['Area'].astype(float)
df['Price'] = df['Price'].astype(float)
df['Bedrooms'] = df['Bedrooms'].astype(float)
df['Bathrooms'] = df['Bathrooms'].astype(float)
```

```python
df['Area']=df['Area'].fillna(df['Area'].mean())
df['Price']=df['Price'].fillna(df['Price'].mean())
df['Bedrooms']=df['Bedrooms'].fillna(df['Bedrooms'].mean())
df['Bathrooms']=df['Bathrooms'].fillna(df['Bathrooms'].mean())

```

```python
df.dropna()
```

```python
df.isnull().sum()
```

```python
df1['bmi'] = df1['bmi'].astype(float)
df1['charges'] = df1['charges'].astype(float)
df1['children'] = df1['children'].astype(float)
```

```python
df1['bmi']=df1['bmi'].fillna(df1['bmi'].mean())
df1['charges']=df1['charges'].fillna(df1['charges'].mean())
df1['children']=df1['children'].fillna(df1['children'].mean())
```

```python
df1.dropna()
```

```python
df1['gender']= label_encoder.fit_transform(df1['gender'])
df1['gender'].unique()
df1['smoker']= label_encoder.fit_transform(df1['smoker'])
df1['smoker'].unique()
```

```python
df1['gender'] = df1['gender'].astype(float)
df1['smoker'] = df1['smoker'].astype(float)
```

```python
df1['gender']=df1['gender'].fillna(df1['gender'].mean())
df1['smoker']=df1['smoker'].fillna(df1['smoker'].mean())
```
